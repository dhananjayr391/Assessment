# LABEL ENCODING

```python
# To convert object data type into numeric data type
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
def label_encoder(dataframe):
    for i in dataframe.columns:
        if dataframe[i].dtypes=="object" or dataframe[i].dtypes=="bool":
            dataframe[i]=le.fit_transform(dataframe[i])
    return dataframe
```

```python
car_updated= label_encoder(car)
car_updated
```

# DETECTING OUTLIERS

```python
def visualize_outliers(dataframe):
    for i in dataframe.columns:
        if dataframe[i].dtypes=="float64" or dataframe[i].dtypes=="int64":
            sns.boxplot(y=i, data=car_updated)
            plt.show()
    return dataframe
```

```python
car_updated1= visualize_outliers(car_updated)
car_updated1
```

### from above we can visualize outliers, inorder to remove outliers we use zscore method.

### Removing Outliers Using zscore Method

```python
from scipy import stats
from scipy.stats import zscore
import numpy as np
```

```python
def remove_outliers(dataframe):
    Z_S= np.abs(zscore(dataframe))
    updated_data= dataframe[(Z_S<3).all(axis=1)]
    return updated_data
```

```python
car_updated2= remove_outliers(car_updated)
car_updated2
```

## Scaling Of Data Using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
mms= MinMaxScaler()
```

```python
def scaling_data(dataframe):
    array_data=mms.fit_transform(dataframe)
    dataframe= pd.DataFrame(array_data, columns=dataframe.columns)
    return dataframe
```

```python
car_updated3= scaling_data(car_updated2)
car_updated3
```

# TRAINING AND TESTING OF DATA

```python
from sklearn.model_selection import train_test_split
def split(dataframe):
    feature_col= dataframe.drop(["Selling_Price"],axis=1)
    target_col= dataframe["Selling_Price"]
    x_train,x_test,y_train,y_test= train_test_split(feature_col,target_col,test_size=0.2,random_state=0,shuffle=True)
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test= split(car_updated3)
x_train
```

# RANDOM FOREST REGRESSOR

# SUPPORT VECTOR REGRESSION

### SVM regression or Support Vector Regression (SVR) is a machine learning algorithm used for regression analysis. It is different from traditional linear regression methods as it finds a hyperplane that best fits the data points in a continuous space, instead of fitting a line to the data points. SVM regression is a powerful tool for regression tasks where there may be complex relationships between the input variables and the target variable.

# MODEL BUILDING

```python
from sklearn.svm import SVR
vector= SVR(kernel = 'rbf')
```

# MODEL TRAINING

```python
vector.fit(x_train,y_train)
```

# PREDICTIONS

```python
y_pred3= vector.predict(x_test)
y_pred3
```

### COMPARING ACTUAL AND PREDICTED SELLING PRICE

```python
compare3= pd.DataFrame({"Actual_selling_price": y_test, "Predicted_selling_price": y_pred3})
compare3
```

# MODEL EVALUATION

```python
# R-Square score
D=r2_score(y_test,y_pred3)
D
```
